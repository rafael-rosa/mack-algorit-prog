```python
#Bubble sort nao otimizado
def bubble_sort(vetor):
    size = len(vetor)
    for i in range(size -1):
      for j in range (size -1):
          if vetor[j] > vetor[j + 1]:
              vetor[j], vetor[j + 1] = vetor[j + 1], vetor[j]
    return vetor
 
#Bubble sort otimizado
def bubble_sort_otim(vetor):
  size = len(vetor)
  for i in range(size -1):
      swapped = False
      for j in range (size -i -1):
          if vetor[j] > vetor[j + 1]:
              vetor[j], vetor[j + 1] = vetor[j + 1], vetor[j]
              swapped = True
      if not swapped:
          break

#Selecao
def selection_sort(vetor):
  for ultimo_slot in range(len(vetor)-1,0,-1):
    position_of_max=0
    for location in range(1,ultimo_slot+1):
      if vetor[location]>vetor[position_of_max]:
        position_of_max = location

    temp = vetor[ultimo_slot]
    vetor[ultimo_slot] = vetor[position_of_max]
    vetor[position_of_max] = temp

#Insercao
def insertion_sort(vetor):
  for index in range(1,len(vetor)):

    current_value = vetor[index]
    position = index

    while position>0 and vetor[position-1]>current_value:
      vetor[position]=vetor[position-1]
      position = position-1

    vetor[position]=current_value

#Mergesort
def merge_sort(vetor):
  # print("Splitting ",vetor)

  if len(vetor)>1:
    mid = len(vetor)//2
    left_half = vetor[:mid]
    right_half = vetor[mid:]

    merge_sort(left_half)
    merge_sort(right_half)

    i=0
    j=0
    k=0
    while i < len(left_half) and j < len(right_half):
      if left_half[i] < right_half[j]:
          vetor[k]=left_half[i]
          i=i+1
      else:
          vetor[k]=right_half[j]
          j=j+1
      k=k+1

    while i < len(left_half):
      vetor[k]=left_half[i]
      i=i+1
      k=k+1

    while j < len(right_half):
      vetor[k]=right_half[j]
      j=j+1
      k=k+1
  # print("Merging ",vetor)


#Quicksort
def quick_sort(alist):
  quick_sort_helper(alist,0,len(alist)-1)

def quick_sort_helper(alist,first,last):
  if first<last:
    split_point = partition(alist,first,last)

    quick_sort_helper(alist,first,split_point-1)
    quick_sort_helper(alist,split_point+1,last)


def partition(alist,first,last):
  pivot_value = alist[first]

  left_mark = first+1
  right_mark = last

  done = False
  while not done:

       while left_mark <= right_mark and alist[left_mark] <= pivot_value:
           left_mark = left_mark + 1

       while alist[right_mark] >= pivot_value and right_mark >= left_mark:
           right_mark = right_mark -1

       if right_mark < left_mark:
           done = True
       else:
           temp = alist[left_mark]
           alist[left_mark] = alist[right_mark]
           alist[right_mark] = temp

  temp = alist[first]
  alist[first] = alist[right_mark]
  alist[right_mark] = temp

  return right_mark
```
