```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(rc={'figure.figsize':(15,8.27)})
```

```python
def convert_to_seconds(time):
    miliseconds = "0." + time.split('.')[1]
    seconds = time.split('.')[0].split(':')[2]
    minutes = time.split('.')[0].split(':')[1]
    hours = time.split('.')[0].split(':')[0]
    seconds = int(hours) * 3600 + int(minutes) * 60 + int(seconds) + (round(float(miliseconds),3))
    return seconds


# Read the CSV file
medicoes_data = pd.read_csv("tempos_execucao_algoritmos.csv")

medicoes_data['segundos_execucao'] = medicoes_data['tempo_execucao'].apply(convert_to_seconds)


# medicoes_data['grupo_tipo_quantidade'] = medicoes_data['tamanho_vetor'].astype(str)+" ITENS "+medicoes_data['tipo_dados']

# View the first 5 rows
medicoes_data.head()
```

```python
medicoes_1000_aleat = medicoes_data.query('tamanho_vetor == 1000 and tipo_dados == "ALEATORIOS"')
medicoes_1000_cresc = medicoes_data.query('tamanho_vetor == 1000 and tipo_dados == "ORDEM_CRESC"')
medicoes_1000_decresc = medicoes_data.query('tamanho_vetor == 1000 and tipo_dados == "ORDEM_DECRESC"')

medicoes_10000_aleat = medicoes_data.query('tamanho_vetor == 10000 and tipo_dados == "ALEATORIOS"')
medicoes_10000_cresc = medicoes_data.query('tamanho_vetor == 10000 and tipo_dados == "ORDEM_CRESC"')
medicoes_10000_decresc = medicoes_data.query('tamanho_vetor == 10000 and tipo_dados == "ORDEM_DECRESC"')

medicoes_100000_aleat = medicoes_data.query('tamanho_vetor == 100000 and tipo_dados == "ALEATORIOS"')
medicoes_100000_cresc = medicoes_data.query('tamanho_vetor == 100000 and tipo_dados == "ORDEM_CRESC"')
medicoes_100000_decresc = medicoes_data.query('tamanho_vetor == 100000 and tipo_dados == "ORDEM_DECRESC"')

medicoes_bubble_sort = medicoes_data.query('algoritmo == "BUBBLE_SORT"')
medicoes_bubble_sort_otim = medicoes_data.query('algoritmo == "BUBBLE_SORT_OTIMIZADO"')
medicoes_selection_sort = medicoes_data.query('algoritmo == "SELECAO"')
medicoes_insertion_sort = medicoes_data.query('algoritmo == "INSERCAO"')
medicoes_quick_sort = medicoes_data.query('algoritmo == "QUICK_SORT"')
medicoes_merge_sort = medicoes_data.query('algoritmo == "MERGE_SORT"')

# medicoes_1000.head(100)
```
